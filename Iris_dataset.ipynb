```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset["data"], iris_dataset["target"], test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_train.shape)
```

```python
df_iris = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
df_iris.head()
```

```python
import mglearn
pd.plotting.scatter_matrix(df_iris, c=y_train, figsize=(15, 15),
marker='o', hist_kwds={'bins': 20}, s=60,
alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
print("The test set accuracy is: ", np.mean(y_pred == y_test))
print("Secnd check ", knn.score(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state = 0)
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```
